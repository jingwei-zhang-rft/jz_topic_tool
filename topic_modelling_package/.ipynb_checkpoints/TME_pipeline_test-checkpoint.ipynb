```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from all_in_one_topic_modelling import PipelineTopicModeling 
from importlib import reload
%load_ext autoreload
%autoreload 2
```

```python
path = '/Users/jingweizhang/Documents/Topic_Modelling/P1_News/'
test_df = pd.read_csv(path + 'trump_trade_corpus.csv')
```

```python
test_df
```

```python
texts = test_df.clean_body.tolist()
```

```python
texts[:2]
```

```python
pltm_object = PipelineTopicModeling(text_data = texts, 
                                    n_gram=2,
                                    tfcounter_type='bow',
                                    model_type='lda',
                                    topic_range_set = [4,10,1])
```

```python
pltm_object.model_saved
```

```python
model_eval,topic_df = pltm_object.get_evaluation_result(coherence_metric = 'c_v', 
                                                          distance_metric = 'cosine', 
                                                          document_score_threshold = 0.5, 
                                                          topn_eval = 20, 
                                                          topn_output = 10)
```

```python
model_eval.to_json()
```

```python
topic_df.iloc[0].top_terms
```

```python
topic_df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
z = model_eval.coherence.tolist()
y = model_eval.distance_avg.tolist()
ax.scatter(z,y)
plt.xlabel('coherence')
plt.ylabel('distance_avg')
marker = model_eval.topic_number.tolist()
for i, txt in enumerate(marker):
    ax.annotate(txt, (z[i], y[i]))
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="coherence", y="distance_avg", hue = 'topic_number', size="document_distribution", data=topic_df)
```
